```python
import requests
from lxml import etree
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
class get_PTT:
    sum_table=[]
    table_pandas={'標題':[],'推數':[],'作者':[],'日期':[],'連結':[]}
    
    def __init__(self,start_page,end_page,to_dataframe=True):
        self.start_page = start_page
        self.end_page = end_page
        self.to_dataframe = to_dataframe
    
    def push_tran(self,string):
        try:
            string_2 = string[0]
        except:
            string_2 = np.nan
        return string_2
    
    def get_table(self,page):
        url='https://www.ptt.cc/bbs/NBA/index{}.html'.format(page)
        header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}
        response = requests.get(url,headers=header)
        html = etree.HTML(response.text)

        title = html.xpath('//div[@class="title"]//a/text()[1]')
        push = [i.xpath('.//text()') for i in html.xpath('//div[@class="nrec"]')]
        author = html.xpath('//div[@class="author"]//text()')
        date = html.xpath('//div[@class="date"]//text()')
        link = ['https://www.ptt.cc/'+ i for i in html.xpath('//div[@class="title"]//a/@href')]
        table = {'標題':title,'推數':push,'作者':author,'日期':date,'連結':link}
        return table
    
    def get_pages(self):
        self.sum_table=[]
        for i in tqdm(range(self.start_page,self.end_page+1)):
            table = self.get_table(i)
            self.sum_table.append(table)
    
    def to_pandas(self):
        self.table_pandas={'標題':[],'推數':[],'作者':[],'日期':[],'連結':[]}
        for i in self.sum_table:
            self.table_pandas['標題'].extend(i['標題'])
            self.table_pandas['推數'].extend(i['推數'])
            self.table_pandas['作者'].extend(i['作者'])
            self.table_pandas['日期'].extend(i['日期'])
            self.table_pandas['連結'].extend(i['連結'])
        self.table_pandas = pd.DataFrame(self.table_pandas)
        self.table_pandas['推數'] = self.table_pandas['推數'].apply(self.push_tran)
        
    def run(self):
        self.get_pages()
        print('='*50)
        print('getting pages succeeded')
        if self.to_dataframe:
            self.to_pandas()
            print('to dataframe succeeded')
```

```python
data=get_PTT(1,100,to_dataframe=True)
data.run()
```
